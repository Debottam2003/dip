```python
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)

img[:] = 255 - img
# Invert the grayscale image(White to Black and Black to White)

cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)

img[:] = img + 50

cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)

img[:] = img - 50

cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)

print(img.shape)

c = 45

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i, j] = c * math.log(1 + img[i, j])

cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Power-Law or Gamma Transformation
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)
print(img.shape)
gamma = 2#Darken the image
#gamma = 0.5#Lighten the image
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i, j] = 1 * (img[i, j] ** gamma)
cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Thresholding Transformation
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)
print(img.shape)
thresh = 128
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i, j] < thresh:
            img[i, j] = 0
        else:
            img[i, j] = 255
cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('thresholded_cloudflare.png', img)
```

```python
#Gray level Slicing
img = cv2.imread('cloudflare.png', cv2.IMREAD_GRAYSCALE)
print(img.shape)
thresh1 = 100
thresh2 = 200
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if thresh1 <= img[i, j] <= thresh2:
            img[i, j] = 255
        else:
            img[i, j] = 0
cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
